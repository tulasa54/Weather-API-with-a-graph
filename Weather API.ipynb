```python
import requests
import json
import matplotlib.pyplot as plt
```

```python
API_key = 'd782bd64d7bd7968444fc28b596ece25'

city_names = ['New York', 'Seattle', 'Miami', 'Chicago', 'Los Angeles', 'Las Vegas', 'London']
temperature_list = []
humidity_list = []

for cities in city_names:
    url = f'https://api.openweathermap.org/data/2.5/weather?q={cities}&appid={API_key}&units=imperial'
    r = requests.get(url)    
    data = r.json()
    
    #print(f'Unformattes JSON:\n{data}\n')
    
    #formatted_json = json.dumps(data, sort_keys = True, indent = 4)
    #print(f'Formatted JSON:\n{formatted_json}')
    
    temperature = data['main']['temp']
    temperature_list.append(temperature)
    humidity = data['main']['humidity']
    humidity_list.append(humidity)
    description = data['weather'][0]['description']
    name = data['name']

    print(f'{name} temperature is {temperature}.')
    print(f'{name} humidity is {humidity}.')
    print(f'{name} weather description is {description}.')
    print()
```

```python
print(city_names)
print(temperature_list)
print(humidity_list)
```

```python
plt.title('Temperature and Humidity for Cities', fontsize = 18)

plt.plot(city_names, temperature_list, color='b', label = 'Temperature', linewidth= 3)
plt.plot(city_names, humidity_list, color='r', label = 'Humidity', linewidth=3)

plt.xlabel('Cities', fontsize = 13)
plt.xticks(rotation = 60)
plt.legend(loc=3)
plt.rcParams['figure.figsize'] = [8,8]
plt.show()
```
